```python
class student: 
    def __init__(self, index, timeStamp, USN): 
        self.index = index
        self.timeStamp = timeStamp 
        self.USN = USN
        
    def print_(self):
        print(self.index,self.timeStamp,self.USN)
        
    def returnUsn(self):
        return self.USN
    
    def returnIndex(self):
        return self.index
    
    def changeIndex(self,newIndex):
        self.index=newIndex
    
    def returnTS(self):
        return self.timeStamp
    
list_of_lists = []
for i in range(0,48):
    list_of_lists.append([])
print('List of lists:')
print(list_of_lists)
```

```python
def PrimaryAllocation():
    x=0
    temp=sorted_data['Choice 1']
    print(temp[0])
    # remove hardcoded value of total students
    for i in range(0,1265):
        list_of_lists[temp[x]].append(student(x,sorted_data['Timestamp'][x],sorted_data['USN'][x]))
        x+=1
```

```python
def CheckElectiveCount():
    global total_count
    total_count=[]
    sum=0
    for i in range(0,48):
        total=0
        for ele in range(0, len(list_of_lists[i])):
            total +=1
        total_count.append(total)
        sum=sum+total
    print(total_count,"Sum : ",sum)
```

```python
def OverflowAllocation():
    global reallocationQueue
    reallocationQueue=[]
    for i in range(0,48):
        if( total_count[i]>under_list[i][1] ):
            for j in range(under_list[i][1],total_count[i]):
                reallocationQueue.append(list_of_lists[i][j])
            del list_of_lists[i][under_list[i][1]:total_count[i]]

    reallocationQueue.sort(key=lambda x: x.timeStamp)
    print(len(reallocationQueue))
```

```python
def UnderflowAllocation():
    global reallocationQueue
#     reallocationQueue=[]
    for i in range(0,48):
        if( total_count[i]<under_list[i][0] ):
            for j in range(0,total_count[i]):
                reallocationQueue.append(list_of_lists[i][j])
            del list_of_lists[i][0:total_count[i]]

    reallocationQueue.sort(key=lambda x: x.timeStamp)
    print(len(reallocationQueue))
```

```python
def LookUpList(usn):
    x=sorted_data.loc[sorted_data['USN'] == usn]
    print(x)
```

```python
def returnChoice(index,choiceNo) -> int:
    return sorted_data[choices[choiceNo-1]][index]
```

```python
def NextAllocation(choiceNo):
    # remove hardcoded value of total students
    for i in range(0,len(reallocationQueue)):
        index_realloc=reallocationQueue[i].returnIndex()
        list_of_lists[returnChoice(index_realloc,choiceNo)].append(reallocationQueue[i])
```

```python
def FinalAllocation(choiceNo):
    for i in range(0,len(reallocationQueue)):
        print(reallocationQueue[i].returnIndex())
    # remove hardcoded value of total students
    for i in range(0,len(reallocationQueue)):
        print('i',i)
        print('len',len(reallocationQueue))
        index_realloc=reallocationQueue[i].returnIndex()
        count=len(list_of_lists[returnChoice(index_realloc,choiceNo)])
        if ( count>=under_list[i][0] and count<under_list[i][1] ):
            list_of_lists[returnChoice(index_realloc,choiceNo)].append(reallocationQueue[i])
            print(index_realloc)
            print('length',len(reallocationQueue))
            del reallocationQueue[index_realloc]
            print(len(reallocationQueue))
```

```python
def SortFinalLists():
    for i in range(0,48):
        list_of_lists[i].sort(key=lambda x: x.timeStamp)
```

```python
def get_key(val):
    for key, value in unique_electives.items():
         if val == value:
             return key

    return "key doesn't exist"
```
